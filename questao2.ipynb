```python
import pandas as pd
import numpy as np
from collections import OrderedDict
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
%matplotlib inline 
from utils import Dataset
```

```python
ds =  Dataset()
data = ds.get_data(['campaign','previous','poutcome', 'y'])

# Inserir coluna para totalizar os contatos realizados
tt_contacts = data.campaign+data['previous'] 
data.insert(loc=2, column='tt_contacts', value=tt_contacts)
```

```python
freq_campaign = data.campaign.value_counts(sort=True)
freq_previous = data.previous.value_counts(sort=True)
freq_tt = data.tt_contacts.value_counts(sort=True)

freq_campaign_relative = data.campaign.value_counts(sort=True, normalize=True)
freq_previous_relative = data.previous.value_counts(sort=True, normalize=True)
freq_tt_relative = data.tt_contacts.value_counts(sort=True, normalize=True)
```

```python
# Min and Max continuous variables:
min_contacts = data.tt_contacts.min()
max_contacts = data.tt_contacts.max()
media_aritimetica = (min_contacts+max_contacts)/2

# Criar var categorica a partir do total dos contatos
qs=[.2,.4,.6,.8,1]
labels=['1','2','3','4']
tt_contacts_categ = pd.qcut(data['tt_contacts'],qs,labels=labels)
data.insert(loc=3,column='tt_contacts_categ', value=tt_contacts_categ)
```
