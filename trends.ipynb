```python
###
# This code answer 2 questions:
# 1 - what profession with a greater tendency to make a loan
# 2 - What most frequent type of loan in the trend profession
# By Sidon 2018
###
```

```python
import pandas as pd
from utils import Dataset
```

```python
ds =  Dataset()
data = ds.get_data(['job','housing','loan'])
```

```python
data
```

```python
# Helper to insert had_loan column
def set_had_loan(row):
    return 'yes' if (row['housing']=='yes' or row['loan']=='yes') else 'no'  
```

```python
# Insert column to indicate if had loan
had_loan = data.apply(set_had_loan , axis=1)
data = pd.concat([data,had_loan], axis=1)
data.columns.values[3]='had_loan'
```

```python
data
```

```python
# Count records in 'housing', 'loan', 'had_loan', grouped by job
count_data_records = data.groupby('job')['housing', 'loan', 'had_loan'].apply(lambda x: x[x=='yes'].count())

# Record count in each job (using housing column) 
data_tt = pd.DataFrame(data=data.groupby(['job']).count()['housing']) 
data_tt.columns = ['tt']
data_tt = pd.concat([data_tt, count_data_records], axis=1)
data_tt['perc_housing']=data_tt.housing/data_tt.tt
data_tt['perc_loan']=data_tt.loan/data_tt.tt
data_tt['perc_tt']=data_tt.had_loan/data_tt.tt
```

```python
data_tt
```

```python
job_trend = data_tt['perc_tt'].idxmax()
record_job_trend = data_tt.loc[job_trend]
perc_job_trend = max(record_job_trend.loan, record_job_trend.housing)
loan_trend = 'Loan' if record_job_trend.loan > record_job_trend.housing else 'Housing'
```

```python
print('Profession with greater tendency to make a loan:', job_trend)
print('Most frequent type of loan in the trend profession:', loan_trend)
```
